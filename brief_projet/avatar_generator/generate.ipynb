```python
import keras
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
generator = keras.models.load_model('/Users/mosbahhachem/Downloads/generator.h5')
```

```python
generator
```

```python
generator.compile()
```

```python
batch_size = 10
latent_dim = 128
random_latent_vectors = tf.random.normal(shape=(batch_size, latent_dim))
```

```python
generated_images = generator.predict(random_latent_vectors)
```

```python
generated_images.shape
```

```python
type(generated_images)
```

```python
import numpy as np
fig = plt.figure(figsize=[16, 8])
for x in range(batch_size):
    ax = plt.subplot(4, 8, x+1)
    plt.imshow((generated_images * 255).astype(np.uint8)[x])
    plt.axis("off")
```
