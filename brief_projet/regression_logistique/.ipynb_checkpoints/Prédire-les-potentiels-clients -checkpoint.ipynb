## 3. Rechagement les données sur la BDD mysql 

```python
from sqlalchemy import create_engine 
import pandas as pd
engine = create_engine("mysql+pymysql://hachem:Tiger13#@localhost/potentielsClients")

def chargement_db(link,table):
    print("Lecture des données")
    csize = 1000
    df = pd.read_csv(link,chunksize = csize, skiprows =0)
    i = csize
    for chunk in df:
        chunk.to_sql(table, con = engine, if_exists= 'append', index = False)
        i += csize
        print(i)
    return print("fin")

chargement_db('./data.csv', 'clients')
```
