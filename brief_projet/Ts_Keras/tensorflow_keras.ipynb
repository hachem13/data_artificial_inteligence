### Simple deep learning with Tensorflow and Keras

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
# Parameters (y = a*x + b)
a=0.6
b=2

# Create noisy data
x_data = np.linspace(-10, 10, num=100000)
y_data = a * x_data + b + np.random.normal(size=100000)
```

```python
# Create the model 
model = keras.Sequential()
model.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model.compile(loss='mse', optimizer="adam")

# Display the model (only 2 parameters to optimize)
model.summary()
```

```python
# Learn
history = model.fit( x_data, y_data, epochs=5, verbose=1)
```

```python
# Predict (compute) the output 
y_predicted = model.predict(x_data)
```

```python
# Display the result
plt.scatter(x_data[::500], y_data[::500])
plt.plot(x_data, y_predicted, 'r', linewidth=4)
plt.grid()
plt.show()
```

```python
print( model.trainable_variables )
```

```python
# Get training and test loss histories
training_loss = history.history['loss']


# Create count of the number of epochs
epoch_count = range(1, len(training_loss) + 1)

# Visualize loss history
plt.plot(epoch_count, training_loss, 'r--')
plt.legend(['Training Loss', 'Test Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show();
```

### Regression no linear with deeplearning

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math
```

```python
# Create noisy data
x_data1 = np.linspace(-10, 10, num=1000)
y_data1 = 0.1*x_data1*np.cos(x_data1) + 0.1*np.random.normal(size=1000)
print('Data created successfully')
```

```python
# Create the model 
model1 = keras.Sequential()
model1.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model1.add(keras.layers.Dense(units = 64, activation = 'relu'))
model1.add(keras.layers.Dense(units = 64, activation = 'relu'))
model1.add(keras.layers.Dense(units = 1, activation = 'linear'))
model1.compile(loss='mse', optimizer="adam")

# Display the model
model1.summary()
```

```python
# Training
history1 = model1.fit( x_data1, y_data1, epochs=100, verbose=1)
```

```python
# Compute the output 
y_predicted1 = model1.predict(x_data1)

# Display the result
plt.scatter(x_data1[::1], y_data1[::1])
plt.plot(x_data1, y_predicted1, 'r', linewidth=4)
plt.grid()
plt.show()
```

### claculate mse

```python
mse = tf.keras.losses.MeanSquaredError()
mse(y_data1, y_predicted1)
```

```python
# Get training and test loss histories
training_loss1 = history1.history['loss']


# Create count of the number of epochs
epoch_count = range(1, len(training_loss1) + 1)

# Visualize loss history
plt.plot(epoch_count, training_loss1, 'r--')
plt.legend(['Training Loss', 'Test Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show();
```

### model 2D 

```python
# Create noisy data
x_data2 = np.linspace(-10, 10, num=1000)
```

```python
x_data3 = np.sin(x_data2) 
X_data4 = x_data2**2

x_data_2D = np.array((x_data3, x_data3)).T

y_data_2D = (x_data3**2 + X_data4**2)**(1/2) + 5*np.random.normal(size = 1000)
```

```python
print(x_data2.shape)
print(x_data_2D.shape)
print(y_data_2D.shape)
```

```python
# Create the model 
model1 = keras.Sequential()
model1.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[2]))
model1.add(keras.layers.Dense(units = 64, activation = 'relu'))
model1.add(keras.layers.Dense(units = 64, activation = 'relu'))
model1.add(keras.layers.Dense(units = 64, activation = 'relu'))
model1.add(keras.layers.Dense(units = 1, activation = 'linear'))
model1.compile(loss='mse', optimizer="adam")

# Display the model
model1.summary()
```

```python
history2 = model1.fit( x_data_2D, y_data_2D, epochs=200, verbose=1)
```

```python
y_predicted2 = model1.predict(x_data_2D)
```
